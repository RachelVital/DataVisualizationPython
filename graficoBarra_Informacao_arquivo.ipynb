```
#Gráfico de Crescimento da População Brasileira
#Lendo informação de um arquivo

import matplotlib.pyplot as plt


dados = open("original.csv").readlines()

x = [] 
y = []

for i in range(len(dados)):
  if i!=0:
    linha = dados[i].split(";")
    x.append(int(linha[0]))
    y.append(int(linha[1]))

# Gráfico de linha
plt.plot(x, y, linestyle = "--")

#Gráfico de Barras
plt.bar(x, y, color = "#e6e6e6")

plt.title("Crescimento da População Brasileira 1980 - 2016")

plt.xlabel("Ano")
plt.ylabel("População x 100.000.000")
plt.show();

# Para salvar
#plt.savefig("populacao_brasileira.png")
```
